W przedsiębiorstwie „W” w Poznaniu przeprowadzono we wrześniu 2022 r. badanie wydajności pracy robotników (mierzonej w kg/godz.). Dla zatrudnionych 15 osób otrzymano następujące wydajności podane w tabeli efficiency_data.
1. Zdefiniuj zbiorowość, jednostkę i badaną cechę statystyczną,
2. Przeprowadź kompleksową analizę struktury robotników zatrudnionych w badanym przedsiębiorstwie według wydajności pracy.


Zbiorowością statystyczną są pracownicy przedsiębiorstwa W w Poznaniu we wrześniu 2022 roku.
Jednostką jest robotnik przediśebiorstwa w W w Poznaniu we wrześniu 20222 roku.
Badaną cechą statystyczną jest efektywność pracwioników W w Poznaniu we wrześniu 2022 roku.

```python
efficiency = [17, 17, 15, 20, 17, 18, 20, 17, 16, 18, 19, 19, 17, 21, 16]
efficiency_df=pd.DataFrame({"efficiency": efficiency})
```

```python
efficiency_df['efficiency'].sum()/len(efficiency_df["efficiency"])
```

Średnia: Przeciętna wydajnośc pracy robotników przedsiębiortswa W w Poznaniu w 2022 roku wynosiła 17,8 kg/h

```python
efficiency_avg = np.mean(efficiency_df['efficiency'])
```

```python
def compute_median(values):
    values_sorted = sorted(values);
    length = len(values_sorted);
    if length % 2 == 0:
        return (values_sorted[length / 2 - 1] + values_sorted[length / 2]) / 2
    return values_sorted[int((length + 1) / 2) -1]
```

```python
compute_median(efficiency)
```

```python
np.median(efficiency_df['efficiency'])
```

Mediana: 50% robotników miało wydajność pracy mniejszą niż 17 kg/h i 50% robotników miało wydajnośc równą lub większą od 50%

```python
np.quantile(efficiency_df['efficiency'], 0.25)
```

25% robotników miało wydajnośc pracy równą lub mniejszą 17 kg/h

```python
np.quantile(efficiency_df['efficiency'], 0.75)
```

3 kwartyl: 75% robotików miało wydajność mniejszą lub równą 19kg/h, a 25% robotników miało wydajność większa lub równą 19 kg/h

```python
efficiency_df['efficiency'].value_counts().idxmax()
efficiency['efficiency'].mode()
```

Dominanta: Wśrod pracowników najczęsciej wystepujaca wydajnoscia jest 17 kg/h

```python
#odchylenie standardowe
efficiency_df["efficiency_diff"] = efficiency_df["efficiency"] - efficiency_avg
efficiency_df["efficiency_diff_power_2"] = efficiency_df["efficiency_diff"] ** 2
std = (efficiency_df["efficiency_diff_power_2"].sum() / (len(efficiency_df['efficiency'])))**0.5
print(std)
```

Odchylenie standardowe: Wydajność robotników przedsiębiorstwa W w Poznnaiu różniła się przeciętnie o 1,64 kg/h od średniej

```python
#współczynnik zmienności - ile procent średniej stanowi odchylenie standardowe
std/efficiency_avg*100
```

```python
#skośność
efficiency_df['efficiency_diff_power_3'] = efficiency_df['efficiency_diff']**3
skewness = efficiency_df['efficiency_diff_power_3'].sum() / (len(efficiency_df['efficiency']) * std**3)
print(skewness)
```

Prawostronnie skośny bo skweness > 0

```python
efficiency_df['efficiency_diff_power_4'] = efficiency_df['efficiency_diff']**4
kurtosis = efficiency_df['efficiency_diff_power_4'].sum() / (len(efficiency_df['efficiency']) * std**4)
print(kurtosis)
```

Kurtoza mniejsza od 3 więc wykres jest spłaszczony
