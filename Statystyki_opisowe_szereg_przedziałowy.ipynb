```python
#srednia
surface_lower_limit = [30, 40, 50, 60, 70, 80, 90]
surface_upper_limit = [40, 50, 60, 70, 80, 90, 100]
number_of_flats = [15, 26, 45, 34, 20, 13, 7]
surface_df = pd.DataFrame({'lower_limit': surface_lower_limit, 'upper_limit': surface_upper_limit, 'number_of_flats': number_of_flats})
surface_df['interval_center'] = (surface_df['lower_limit'] + surface_df['upper_limit'] ) / 2
surface_avg = sum(surface_df['number_of_flats']*surface_df['interval_center'])/surface_df['number_of_flats'].sum()
```

```python
#odchylenie standardowe
surface_df['surface_diff'] = surface_df['interval_center'] - surface_avg
surface_df['surface_diff_power2'] = surface_df['surface_diff']**2
surface_std = (sum(surface_df['surface_diff_power2']*surface_df['number_of_flats'])/surface_df['number_of_flats'].sum())**0.5
surface_std
```

```python
#wspolczynnik zmiennosci
surface_std/surface_avg
```

```python
#skosnosc
surface_df['surface_diff_power3'] = surface_df['surface_diff']**3
skewness = sum(surface_df['surface_diff_power3']*surface_df['number_of_flats'])/surface_df['number_of_flats'].sum()/surface_std**3
skewness
```

```python
#kurtoza
surface_df['surface_diff_power4'] = surface_df['surface_diff']**4
kurtosis = sum(surface_df['surface_diff_power4']*surface_df['number_of_flats'])/surface_df['number_of_flats'].sum()/surface_std**4
kurtosis
```
