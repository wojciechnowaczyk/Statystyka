Przeprowadzono badanie pracowników pewnej firmy w Poznaniu w lipcu 2022 r. według liczby dzieci na utrzymaniu. Zebrane dane pogrupowano i zbudowano szereg rozdzielczy w ramce kids_df.
W oparciu o podane informacje:
1. Określ zbiorowość, jednostkę i badaną cechę statystyczną.
2. Przeprowadź kompleksową analizę struktury pracowników badanej firmy według liczby dzieci na utrzymaniu

```python
dependent_children = [0, 1, 2, 3, 4, 5, 6]
number_of_employees = [9, 16, 26, 10, 5, 3, 1]
kids_df = pd.DataFrame({'kids': dependent_children, 'employees_no': number_of_employees})
kids_df
```

Zbiorowość - to na kim robimy badanie - pracownicy
Badana cecha: ilość dzieci
Jednostka statystyczna: pracownik

```python
#srednia
kids_df['kids_total'] = kids_df["kids"]*kids_df["employees_no"]
kids_avg = kids_df['kids_total'].sum()/kids_df["employees_no"].sum();
```

```python
#dominanta
print(kids_df[kids_df['employees_no']== kids_df['employees_no'].max()]['kids'].iloc[0]);
```

```python
#mediana
kids_df['employees_cum'] = kids_df['employees_no'].cumsum();
employees_no = kids_df['employees_no'].sum()
if employees_no % 2 == 0:
    median = (kids_df[kids_df['employees_cum'] >= employees_no/2]['kids'].iloc[0] + kids_df[kids_df['employees_cum'] >= employees_no/2+1]['kids'].iloc[0])/2
else:
    median = kids_df[kids_df['employees_cum'] >= employees_no/2]['kids'].iloc[0]
```

```python
# odchylenie standardowe
kids_df['kids_diff'] = kids_df['kids'] - kids_avg
kids_df["kids_diff_power_2"] = kids_df["kids_diff"]**2
kids_df['kids_diff_power_2_total'] = kids_df["kids_diff_power_2"]*kids_df["employees_no"]
std = (kids_df['kids_diff_power_2_total'].sum()/kids_df["employees_no"].sum())**0.5
std
```

```python
# wspolczynnik zmiennosci
std/kids_avg
```

```python
#skosnosc
kids_df['kids_diff_power_3'] = kids_df["kids_diff"]**3
kids_df['kids_diff_power_3_total'] = kids_df["kids_diff_power_3"]*kids_df["employees_no"]
skewness = kids_df['kids_diff_power_3_total'].sum()/kids_df["employees_no"].sum()/std**3
skewness
```

```python
# kurtoza
kids_df['kids_diff_power_4'] = kids_df["kids_diff"]**4
kids_df['kids_diff_power_4_total'] = kids_df["kids_diff_power_4"]*kids_df["employees_no"]
kurtosis = kids_df['kids_diff_power_4_total'].sum()/kids_df["employees_no"].sum()/std**4
kurtosis
```
